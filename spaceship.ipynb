```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.impute import SimpleImputer
```

```python
SpendTot=["RoomService","FoodCourt","ShoppingMall","Spa","VRDeck"]
train["SpendTot"]=train[SpendTot].sum(axis=1)
test["SpendTot"]=test[SpendTot].sum(axis=1)
train[['c1','c2','c3']]=train['Cabin'].str.split('/',expand=True)
test[['c1','c2','c3']]=test['Cabin'].str.split('/',expand=True)
```

```python
numeric=train.select_dtypes(include=['number']).columns
categoric=train.select_dtypes(include=['object']).drop(columns=['Name','Cabin','PassengerId']).columns
numeric_transformers=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='mean')),
    ('scaler',StandardScaler())
])
categoric_transformers=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('encoder',OneHotEncoder(handle_unknown='ignore'))
])
preprocessing=ColumnTransformer(transformers=[
    ('num',numeric_transformers,numeric),
    ('cat',categoric_transformers,categoric)
])
pipeline=Pipeline(steps=[
    ('preprocessing',preprocessing),
    ('model',XGBClassifier())
])
```

```python
y=train.Transported
X=train.drop(columns=['Transported','Name','Cabin','PassengerId'])
```

```python


X_train, X_test,y_train,y_test=train_test_split(X,y,random_state=42)
# params={
#     "model__n_estimators":[100,500,1000],
#     "model__learning_rate":[0.01,0.05,0.1],
#     "model__reg_alpha":[0.1,0,1],
#     "model__reg_lambda":[0.1,0,1],
#     "model__max_depth":[3,5,7],
#     "model__subsample":[0.7,0.8,1],
#     "model__colsample_bytree":[0.5,0.7,0.9]
# }
# grid_search=GridSearchCV(
#     estimator=pipeline,
#     param_grid=params,
#     cv=3,
#     scoring="accuracy",
#     verbose=1,
#     n_jobs=3
# )
# grid_search.fit(X_train,y_train)
# best_model=grid_search.best_estimator_
# pred=best_model.predict(X_test)
# print("Best params: ",grid_search.best_params_)
# print("accuracy:", accuracy_score(y_test,pred))


pipeline.set_params(
    model__n_estimators=1000,
    model__max_depth=5,
    model__learning_rate=0.01,
    model__colsample_bytree=0.9,
    model__reg_lambda=0.1,
    model__reg_alpha=1,
    model__subsample=0.8
)
pipeline.fit(X_train,y_train)
pred=pipeline.predict(X_test).astype(bool)
print("accuracy",accuracy_score(y_test,pred))
pred_main=pipeline.predict(test).astype(bool)

```

```python
sub=pd.DataFrame({
    'PassengerId':test.PassengerId,
    'Transported':pred_main
})
sub.to_csv("submission.csv",index=False)
sub
```
